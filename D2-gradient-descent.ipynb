```python
import numpy as np
import pandas as pd

class GradientDescent():
    
    def __init__(self,start_pos,epochs=500,learning_rate=0.001):
        
        self.epochs = epochs
        self.learning_rate = learning_rate
        self.best_guess = []
        
        
    def paraboloid(self,x,y):
        return x**2+y**2
    
    def compute_gradient(self,vector):
        
        if len(vector)==2:
            
            x,y = vector
            
            return [2*x , 2*y]
        
    def compute_step(self,current_pos,learning_rate):
        
        grad = self.compute_gradient(current_pos)
        
        grad[0] *= -learning_rate  # gradient = gradient x  -1(learning_rate)
        grad[1] *= -learning_rate
        
        next_pos = [0,0]
        
        next_pos[0] = current_pos[0] + grad[0]
        next_pos[1] = current_pos[1] + grad[1]
        
        return next_pos
    
    def prediction(self):
        
        
        print(f"Starting position is {start_pos}")
        # inintially best position will be starting position
        
        best_pos=start_pos
        
       
        for i in range(1, self.epochs):
            next_pos = self.compute_step(best_pos,self.learning_rate)
            
            if i % 100 ==0:
                print(f'{i} is and guess is {next_pos}')
            best_pos = next_pos
        self.best_guess = best_pos
        
        
        print("Best Guess for minimum is ",self.best_guess)
        
```

### Minimum of a paraboloid is (0,0)



#### and After 500 epochs we are reaching a Minimum

```python

# seeding numpy
np.random.seed(34)

start_pos = np.random.randint(low=1,high=10,size=(1,2)).tolist()[0]

g=GradientDescent(epochs=500,learning_rate=0.009,start_pos=start_pos)

g.prediction()
```
